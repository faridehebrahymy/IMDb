```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from time import sleep
from typing import List
from typing import Optional
from sqlalchemy import create_engine
from sqlalchemy import URL
from sqlalchemy import create_engine, VARCHAR, Integer
from sqlalchemy.orm import sessionmaker
import mysql.connector
from mysql.connector import errorcode
from datetime import date, datetime, timedelta
from concurrent.futures import ThreadPoolExecutor
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9",
}

url = "https://www.imdb.com/chart/top/?ref_=nv_mv_250"
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
movies = list(soup.select("td.titleColumn a"))

def get_movie_data(movie):
    movie_id1 = movie["href"].split("/")[2]
    movie_id = movie["href"].split("/")[2].replace("tt", "")
    movie_url = f"https://www.imdb.com/title/{movie_id1}"
    #count+=1
    #print(count)
    try:
        response = requests.get(movie_url, headers=headers)
        response.raise_for_status()
        movie_soup = BeautifulSoup(response.text, "html.parser")

        title_element = movie_soup.find("span", class_="sc-afe43def-1 fDTGTb")
        title = title_element.text.strip() if title_element is not None else ""

        ul_element = movie_soup.find("ul", class_="ipc-inline-list ipc-inline-list--show-dividers sc-afe43def-4 kdXikI baseAlt")
        lis = ul_element.find_all("li", class_="ipc-inline-list__item")
        list_a = ul_element.find_all("a", class_="ipc-link ipc-link--baseAlt ipc-link--inherit-color")
        year = int(list_a[0].text.strip()) if len(lis) > 0 else 0
        parental_guide = list_a[1].text.strip() if len(list_a) > 1 else ""
        runtime_element = lis[-1].text if len(lis) > 2 else None
        if movie_id=="0082096":
            parental_guide=""
            runtime=149
        if runtime_element:
            tokens = runtime_element.split()
            runtime = 0
            for i, token in enumerate(tokens):
                if token.endswith("h"):
                    hours = int(token[:-1])
                    runtime += hours * 60
                elif token.endswith("m"):
                    minutes = int(token[:-1])
                    runtime += minutes


        genres_element = movie_soup.find("div", class_="ipc-chip-list__scroller")
        genres_element1 = genres_element.select(".ipc-chip__text")
        genres = [genre.text.strip() for genre in genres_element1] if genres_element1 else []


        ul_element1 = movie_soup.find("ul", class_="ipc-inline-list ipc-inline-list--show-dividers ipc-inline-list--inline ipc-metadata-list-item__list-content baseAlt")
        directors_elements = ul_element1.find_all("a", class_="ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link")   
        directors = []
        director_ids = []

        for director_element in directors_elements:
            directors.append(director_element.text)
            director_ids.append(director_element["href"].split("/")[2].replace('nm',''))



        writer_element1 = movie_soup.find_all("div",class_="ipc-metadata-list-item__content-container")
        writer_container = writer_element1[1].find_all("li", class_="ipc-inline-list__item")
        writer_elements = []
        for container in writer_container:
            writer_elements.extend(container.find_all("a", class_="ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link"))
        writers = []
        writer_ids = []
        for writer_element in writer_elements:
            writer = writer_element.text.strip()
            writer_id = writer_element["href"].split("/")[2].replace("nm", "")
            writers.append(writer)
            writer_ids.append(writer_id)



        star_element0= movie_soup.find("ul", class_="ipc-metadata-list ipc-metadata-list--dividers-all title-pc-list ipc-metadata-list--baseAlt")
        #print(star_element0)
        star_element= star_element0.find_all("li", class_="ipc-metadata-list__item ipc-metadata-list-item--link")
        #print(star_element)
        star_elements = star_element[-1].find_all("a", class_="ipc-metadata-list-item__list-content-item ipc-metadata-list-item__list-content-item--link")
        #print(star_elements)
        stars = []
        star_ids = []
        for star_element in star_elements:
            star = star_element.text.strip()
            star_id = star_element["href"].split("/")[2].replace("nm", "")
            stars.append(star)
            star_ids.append(star_id)



        li_elements = movie_soup.find_all("li", class_="ipc-metadata-list__item sc-6d4f3f8c-2 byhjlB")
        gross_us_canada = ""
        for li_element in li_elements:
            label_span = li_element.find_all("span", class_="ipc-metadata-list-item__label")
            for label in label_span:
                if label.text == "Gross US & Canada":
                    span_element = li_element.find("span", class_="ipc-metadata-list-item__list-content-item")
                    if span_element is not None:
                        gross_us_canada = span_element.text.replace("$", "").replace(",", "")
        if gross_us_canada == "":
            gross_us_canada = None
        return {"movie_id": movie_id,
                "Title": title,
                "Year": year,
                "ParentalGuide": parental_guide,
                "Runtime": runtime,
                "Genres": genres,
                "Director":director,
                "director_id":director_id,
                "Writers":writers,
                "writer_ids":writer_ids, 
                "Stars":stars,
                "star_ids":star_ids,
                "gross_us_canada": gross_us_canada}
    except requests.exceptions.RequestException as e:
        print(e)
        return None

```

```python
# Split the list of movies into chunks of 50
chunks = [movies[i:i+50] for i in range(0, len(movies), 50)]

data = []  # create an empty list to hold the collected data

with ThreadPoolExecutor(max_workers=10) as executor:  # create a thread pool with a maximum of 10 workers
    for chunk in chunks:  # iterate through each chunk of movies
        futures = [executor.submit(get_movie_data, movie) for movie in chunk]  # the submit() method is used inside a loop to concurrently execute the get_movie_data function for each movie in a given chunk using multiple threads. 
        results = [future.result() for future in futures]  # get the result for each future object
        data.extend([r for r in results if r is not None])  # add non-empty results to the data list
        sleep(3)  # sleep for 3 second to avoid sending too many requests at once


df = pd.DataFrame(data, columns=["movie_id","Title", "Year", "ParentalGuide", "Runtime", "Genres", "Director","director_id", "Writers","writer_ids" ,"Stars","star_ids", "gross_us_canada"])
df
```

```python
def create_database(cursor, DB_NAME):
    try:
        cursor.execute(f"DROP DATABASE IF EXISTS {DB_NAME}")
        cursor.execute(f"CREATE DATABASE IF NOT EXISTS {DB_NAME} DEFAULT CHARACTER SET 'utf8'")
        cursor.reset()
        print(f"Successfully created database: {DB_NAME}")
    except mysql.connector.Error as err:
        print(f"Failed creating database: {err}")
```

# Create Tables

```python
TABLES = {}
# Define the movie table
TABLES['movie'] = (
    "CREATE TABLE IF NOT EXISTS movie ("
    "  movie_id VARCHAR(8) NOT NULL,"
    "  Title VARCHAR(128),"
    "  Year INT,"
    "  ParentalGuide VARCHAR(8),"
    "  Runtime INT,"
    "  gross_us_canada INT,"
    "  PRIMARY KEY (movie_id)"
    ") ENGINE=InnoDB")
# Define the person table
TABLES['person'] = (
    "CREATE TABLE IF NOT EXISTS person ("
    "  person_name VARCHAR(32) NOT NULL,"
    "  person_id VARCHAR(8) NOT NULL,"
    "  PRIMARY KEY (person_id)"
    ") ENGINE=InnoDB")

# Define the genre table
TABLES['genre'] = (
    "CREATE TABLE IF NOT EXISTS genre ("
    "  id INT NOT NULL AUTO_INCREMENT,"
    "  movie_id VARCHAR(8),"
    "  genre VARCHAR(16),"
    "  PRIMARY KEY (id),"
    "  FOREIGN KEY (movie_id) REFERENCES movie(movie_id)"
    ") ENGINE=InnoDB AUTO_INCREMENT=1")


# Define the cast table
TABLES['cast'] = (
    "CREATE TABLE IF NOT EXISTS cast ("
    "  id INT NOT NULL AUTO_INCREMENT,"
    "  movie_id VARCHAR(8),"
    "  person_id VARCHAR(8),"
    "  PRIMARY KEY (id),"
    "  FOREIGN KEY (movie_id) REFERENCES movie(movie_id),"
    "  FOREIGN KEY (person_id) REFERENCES person(person_id)"
    ") ENGINE=InnoDB AUTO_INCREMENT=1")


# Define the crew table
TABLES['crew'] = (
    "CREATE TABLE IF NOT EXISTS crew ("
    "  id INT NOT NULL AUTO_INCREMENT,"
    "  movie_id VARCHAR(8),"
    "  person_id VARCHAR(8),"
    "  Role VARCHAR(8),"
    "  PRIMARY KEY (id),"
    "  FOREIGN KEY (movie_id) REFERENCES movie(movie_id),"
    "  FOREIGN KEY (person_id) REFERENCES person(person_id)"
    ") ENGINE=InnoDB AUTO_INCREMENT=1")

# Set AUTO_INCREMENT to start from 1 for tables with AUTO_INCREMENT column
ALTER_TABLES = [
    "ALTER TABLE genre AUTO_INCREMENT=1",
    "ALTER TABLE cast AUTO_INCREMENT=1",
    "ALTER TABLE crew AUTO_INCREMENT=1"
]

```

```python
cursor.execute(f"USE {DB_NAME}")
for table_name in TABLES:
    table_description = TABLES[table_name]
    try:
        print("Creating table {}: ".format(table_name), end='')
        cursor.execute(table_description)
    except mysql.connector.Error as err:
        if err.errno == errorcode.ER_TABLE_EXISTS_ERROR:
            print("already exists.")
        else:
            print(err.msg)
    else:
        print("OK")

cursor.close()
```

# Person Table

```python
person_names = []
person_id = []
#print(director)
# بررسی هر سطر دیتافریم اصلی
for index, row in df.iterrows():
    # استخراج اطلاعات مربوط به کارگردان
    director = row['Director']
    director_ids = row['director_id']
    # اضافه کردن اطلاعات کارگردان به لیست
    for i in range(len(director)):
        person_names.append(director[i])
        person_id.append(director_ids[i])
    
    # استخراج اطلاعات مربوط به نویسندگان
    writers = row['Writers']
    writer_ids = row['writer_ids']
    # اضافه کردن اطلاعات نویسندگان به لیست
    for i in range(len(writers)):
        person_names.append(writers[i])
        person_id.append(writer_ids[i])
    
    # استخراج اطلاعات مربوط به بازیگران
    stars = row['Stars']
    star_ids = row['star_ids']
    # اضافه کردن اطلاعات بازیگران به لیست
    for i in range(len(stars)):
        person_names.append(stars[i])
        person_id.append(star_ids[i])

# ایجاد دیتافریم جدید "person"
person = pd.DataFrame({'person_name': person_names, 'person_id': person_id})
person.drop_duplicates(subset=['person_id'], keep='first', inplace=True)

# نمایش دیتافریم "person"
person
```

# Cast Table

```python
# Create the 'crew' DataFrame
cast_data = []
for index, row in df.iterrows():
    movie_id = row["movie_id"]
    director_id = row["director_id"]
    writer_ids = row["writer_ids"]
    star_ids = row["star_ids"]

    # Add directors to the crew_data list
    for id in director_id:
        cast_data.append([movie_id, id])

    # Add writers to the crew_data list
    for writer, writer_id in zip(row["Writers"], writer_ids):
        cast_data.append([movie_id, writer_id])

    #Add stars to the crew_data list
    for star, star_id in zip(row["Stars"], star_ids):
        cast_data.append([movie_id, star_id])


# Create the 'crew' DataFrame
cast= pd.DataFrame(cast_data, columns=["movie_id", "person_id"])

cast
```

```python
cnx = mysql.connector.connect(user=user, password=password,
                              host='localhost ',
                              port=3306)
cursor = cnx.cursor()
cursor.execute(f'USE {DB_NAME}')


# Iterate over the rows of the DataFrame and insert each row into the MySQL table
for index, row in cast.iterrows():
    # Replace any NaN values in the row with None
    row = row.where(pd.notnull(row), None)
    
    sql = "INSERT INTO cast (movie_id, person_id) VALUES (%s, %s)"
    val = (row['movie_id'], row['person_id'])
    cursor.execute(sql, val)

cnx.commit()
```
